```
import math
import scipy.stats

S0 = float(input("輸入S0："))
r = float(input("輸入r："))
q = float(input("輸入q："))
Sigma = float(input("輸入Sigma："))
T = float(input("輸入T："))
K1 = float(input("輸入K1："))
K2 = float(input("輸入K2："))
K3 = float(input("輸入K3："))
K4 = float(input("輸入K4："))

def dplus(S0, K, r, q, Sigma, T):
  result = (math.log(S0/K)+(r-q+Sigma**2/2)*T) / (Sigma * T**0.5)
  return result
def dminus(S0, K, r, q, Sigma, T):
  result = (math.log(S0/K)+(r-q-Sigma**2/2)*T) / (Sigma * T**0.5)
  return result

payoff1 = math.exp(-r*T) * S0 * math.exp((r-q)*T) * (scipy.stats.norm.cdf(dplus(S0, K1, r, q, Sigma, T)) - scipy.stats.norm.cdf(dplus(S0, K2, r, q, Sigma, T)))
payoff2 = -K1 * math.exp(-r*T) * (scipy.stats.norm.cdf(dminus(S0, K1, r, q, Sigma, T)) - scipy.stats.norm.cdf(dminus(S0, K2, r, q, Sigma, T)))
payoff3 = (K2 - K1) * math.exp(-r*T) * (scipy.stats.norm.cdf(dminus(S0, K2, r, q, Sigma, T)) - scipy.stats.norm.cdf(dminus(S0, K3, r, q, Sigma, T)))
payoff4 = K4 * ((K2 - K1)/(K4 - K3)) * math.exp(-r*T) * (scipy.stats.norm.cdf(dminus(S0, K3, r, q, Sigma, T)) - scipy.stats.norm.cdf(dminus(S0, K4, r, q, Sigma, T)))
payoff5 = -((K2 - K1)/(K4 - K3)) * math.exp(-r*T) * S0 * math.exp((r-q)*T) * (scipy.stats.norm.cdf(dplus(S0, K3, r, q, Sigma, T)) - scipy.stats.norm.cdf(dplus(S0, K4, r, q, Sigma, T)))

payoff = payoff1 + payoff2 + payoff3 + payoff4 + payoff5
payoff
```
